```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
# set the parameters
r = 0.12
v0 = 0.5
t1 = 10
t2 = 5
tEnd = 0
```

```python
def getVL(t):
    return v0 * (t / t1)
def getVR(t):
    return v0 * (t / t2)
```

```python
def runDynamics(dt = 0.01, tEnd = 10):
    t = 0
    xTracker = []
    yTracker = []
    phiTracker = []
    xDotTracker = []
    yDotTracker = []
    timeTracker = []
    x0 = 0
    y0 = 0
    phi0 = 0

    while t < tEnd:
        t += dt
        timeTracker.append(t)
        xDot = np.cos(phi0) * (getVL(t) + getVR(t)) / 2 
        yDot = np.sin(phi0) * (getVL(t) + getVR(t)) / 2 

        x0 += dt * xDot
        y0 += dt * yDot
        phi0 += dt * (getVR(t) - getVL(t)) / (2 * r)

        xTracker.append(x0)
        yTracker.append(y0)
        xDotTracker.append(xDot)
        yDotTracker.append(yDot)
        phiTracker.append(phi0)

    return xTracker, yTracker, phiTracker, xDotTracker, yDotTracker, timeTracker
```

```python
# visualize the trajectory with a colormap

x, y, phi, xVel, yVel, time = runDynamics()

plt.plot(x, y, label = "trajectory")
plt.legend()
plt.grid()
plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.show()

plt.plot(time,xVel,  label = "xVel")  
plt.plot(time,yVel, label = "yVel")
plt.legend()
plt.xlabel("time [s]")
plt.ylabel("velocity [m/s]")
plt.grid()
plt.show()
```
